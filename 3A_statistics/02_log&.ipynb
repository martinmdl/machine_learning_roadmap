```python
import pandas as pd

df = pd.DataFrame({
    'x': [1, 10, 100, 1000, 10000, 100000, 1000000, 10000000],
    'y': [1, 2, 3, 4, 5, 6, 7, 8]
})
```

```python
df.plot(x='x', y='y', logx=False, logy=False, xlabel='actual magnitude', ylabel='richter scale', kind='line')
df.plot(x='x', y='y', logx=True, logy=True, xlabel='actual magnitude', ylabel='richter scale', kind='line')
df.plot(x='x', y='y', logx=False, logy=False, xlabel='actual magnitude', ylabel='richter scale', kind='bar')
df.plot(x='x', y='y', logx=False, logy=True, xlabel='actual magnitude', ylabel='richter scale', kind='bar')
```
